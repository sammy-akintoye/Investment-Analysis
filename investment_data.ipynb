```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\hp\\Desktop\\csv_folders\\Finance_data.csv")

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

df.columns = df.columns.str.title()

df

df.info()
#Checking the information of the DataSet

df.describe()

#Using a bar plot to find the highest average and lowest average


mean_values = df.describe().loc["mean"]
mean_values = mean_values.drop("Age")

mean_values.plot(kind="bar", figsize=(10, 6), color="skyblue")
plt.title("A Bar Plot of Financial Instruments")
plt.ylabel("Average Values of Investment")
plt.xlabel("Investment Avenue(s)")
plt.xticks(rotation=45)
plt.show()
```
